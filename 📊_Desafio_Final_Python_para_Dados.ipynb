# ‚úÖ **1. Importa√ß√£o das Bibliotecas**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display  # Para exibir tabelas no Jupyter Notebook
```

# ‚úÖ **2. Carregamento da Base de Dados**

```python
file_path ="/content/1KBtJf20QHGyrNOTEGeq_saude_do_sono_estilo_vida.csv"
df = pd.read_csv(file_path)
```

# Exibir as primeiras linhas

```python
print("üìå Visualizando os primeiros registros do dataset:")
display(df.head())
```

# ‚úÖ **3. Renomeando as Colunas**

```python
df.rename(columns={
    "ID": "Identificador",
    "Ocupa√ß√£o": "Profiss√£o",
    "Categoria BMI": "Categoria IMC",
    "Press√£o sangu√≠neaaaa": "Press√£o sangu√≠nea"
}, inplace=True)

print("‚úÖ Colunas renomeadas com sucesso!")
display(df.head())
```

# ‚úÖ **4. Estat√≠sticas do Sono por Profiss√£o**
# M√©dia, Mediana e Moda das horas de sono para cada profiss√£o

```python
estatisticas_sono = df.groupby("Profiss√£o")["Dura√ß√£o do sono"].agg(["mean", "median", lambda x: x.mode()[0]])
estatisticas_sono.rename(columns={"<lambda_0>": "mode"}, inplace=True)

print("üìä Estat√≠sticas do Sono por Profiss√£o:")
display(estatisticas_sono)
```

# ‚úÖ **5. Porcentagem de Obesos entre Engenheiros de Software**

```python
engenheiros = df[df["Profiss√£o"] == "Eng. de Software"]
percentual_obesos = (engenheiros["Categoria IMC"] == "Obesidade").mean() * 100

print(f"üìå Porcentagem de Engenheiros de Software obesos: {percentual_obesos:.2f}%")
```

# ‚úÖ **6. Advogados ou Representantes de Vendas Dormem Menos?**

```python
profissoes_alvo = ["Advogado(a)", "Representante de Vendas"]
df_filtrado = df[df["Profiss√£o"].isin(profissoes_alvo)]
media_sono_profissoes = df_filtrado.groupby("Profiss√£o")["Dura√ß√£o do sono"].mean()

print("üìä M√©dia de Horas de Sono por Profiss√£o:")
display(media_sono_profissoes)
```

# ‚úÖ **7. Quem Dorme Menos: Enfermeiros ou M√©dicos?**

```python
profissoes_medicas = ["Enfermeiro(a)", "M√©dico(a)"]
df_medicos = df[df["Profiss√£o"].isin(profissoes_medicas)]
media_sono_medicos = df_medicos.groupby("Profiss√£o")["Dura√ß√£o do sono"].mean()

print("üìä M√©dia de Horas de Sono: Enfermeiros vs. M√©dicos")
display(media_sono_medicos)
```

# ‚úÖ **8. Criar um Subconjunto com Informa√ß√µes Selecionadas**

```python
subconjunto = df[["Identificador", "G√™nero", "Idade", "Press√£o sangu√≠nea", "Frequ√™ncia card√≠aca"]]

print("üìÇ Subconjunto de Dados Selecionados:")
display(subconjunto.head())
```

# ‚úÖ **9. Profiss√£o Menos Frequente**

```python
profissao_menos_frequente = df["Profiss√£o"].value_counts().idxmin()

print(f"üìå A profiss√£o menos frequente na base de dados √©: {profissao_menos_frequente}")
```

# ‚úÖ **10. Quem tem maior Press√£o Sangu√≠nea M√©dia: Homens ou Mulheres?**

```python
df["Press√£o Sist√≥lica"] = df["Press√£o sangu√≠nea"].apply(lambda x: int(x.split('/')[0]))
pressao_media_genero = df.groupby("G√™nero")["Press√£o Sist√≥lica"].mean()

print("üìä M√©dia da Press√£o Sist√≥lica por G√™nero:")
display(pressao_media_genero)
```

# ‚úÖ **11. Predomin√¢ncia de Sono de 8 Horas**

```python
moda_sono = df["Dura√ß√£o do sono"].mode()[0]

if moda_sono == 8:
    print("‚úÖ Sim, a maioria das pessoas dorme 8 horas por dia.")
else:
    print(f"üìå N√£o, a moda das horas de sono √© {moda_sono} horas.")
```

# ‚úÖ **12. Pessoas com Frequ√™ncia Card√≠aca Acima de 70 D√£o Mais Passos?**

```python
passos_alta_freq = df[df["Frequ√™ncia card√≠aca"] > 70]["Passos di√°rios"].mean()
passos_baixa_freq = df[df["Frequ√™ncia card√≠aca"] <= 70]["Passos di√°rios"].mean()

print(f"üìä Passos M√©dios - Frequ√™ncia Card√≠aca Alta (>70): {passos_alta_freq:.2f}")
print(f"üìä Passos M√©dios - Frequ√™ncia Card√≠aca Baixa (‚â§70): {passos_baixa_freq:.2f}")

if passos_alta_freq > passos_baixa_freq:
    print("‚úÖ Pessoas com frequ√™ncia card√≠aca acima de 70 d√£o mais passos em m√©dia.")
else:
    print("‚ùå Pessoas com frequ√™ncia card√≠aca acima de 70 n√£o d√£o mais passos em m√©dia.")
```
